```python
#3) 대여소정보 join 대여정보
mergeData = pd.merge(place, rent_info, left_on='대여소ID', right_on='대여 대여소번호')
mergeData
```

```python
#4) 각 구별 평균 이용시간, 이용거리
#4-1 각 구별 평균 이용시간
mergeData.groupby("대여소_구")["이용시간"].mean()
```

```python
#4) 각 구별 평균 이용시간, 이용거리
#4-1 각 구별 평균 이용시간
mergeData.groupby("대여소_구")["이용거리"].mean()
```

```python
# 5) "대여소_구" 별로 최대이용시간, 최소이용시간 
mergeDataGroup = mergeData.groupby("대여소_구")["이용시간"]
mergeDataGroup.agg([np.sum,np.mean,np.max,np.min])
```
