```python
df1 = pd.DataFrame(
    {'번호' : [ 20, 30, 40, 50, 60, 70], #번호는 SQL로 치면 PK역할을 해준다.
     '이름' : ['임꺽정', '전우치', '손오공', '저팔계', '사오정', '삼장법사']},
)
df1
```

```python
df2 = pd.DataFrame(
    {'번호' : [10, 10, 20, 40, 50, 40, 40, 70], #출력을 원치 않는 데이터는 삭제하면 된다. 여기선 60번 삭제.
     '금액' : [1000,2500,4500,3400,1100,230,3400,3550]},#60번과 관련된 금액 3311 삭제
)
df2
```

```python
#Merge 실시 예쩨
# 두 DataFrame에서 공통열 (Pk, Fk)이 필요하다.
pd.merge(df1,df2)

#공통열에 해당하는 것들을 출력하긴 하는데 
```

```python
df2 = pd.DataFrame(
    {'번호' : [10, 10, 20, 40, 50, 40, 40, 60, 70], #번호는 SQL로 치면PK역할을 해준다. 그렇다고 df1의 번호랑 같은 애인건 아냐.(#동명이인)
     '금액' : [1000,2500,4500,3400,1100,230,3400,3311,3550]},
)
df2
```

```python
pd.merge(df1,df2, how='outer') #outer join(정확히는 full outer join) # NaN인 것들도 출력된다.
```

```python
pd.merge(df1,df2, how='left') #
```

```python
pd.merge(df1,df2, how='right') #
```

# join on

```python
df1 = pd.DataFrame(
    {'번호' : [10, 20, 30, 40, 50, 60, 70], #번호는 SQL로 치면 PK역할을 해준다.
     '이름' : ['홍길동', '임꺽정', '전우치', '손오공', '저팔계', '사오정', '삼장법사'],
     '타입' : ['우수','일반','우수','우수','일반','일반','우수']
    },
)
df1
```

```python
df2 = pd.DataFrame(
    {'고객번호' : [10, 10, 20, 40, 50, 40, 40, 60, 70], #PK가 '번호', 여기는 '고객번호'로 공통컬럼명 없기에 에러난다.
     '금액' : [1000,2500,4500,3400,1100,230,3400,3311,3550],    
    },
)
df2
```

```python
pd.merge(df1,df2) # 공통컬럼명이 없어서 에러나는 예제
```

```python
#공통컬럼명이 없어도 문제없게 출력되도록 한다.
pd.merge(df1,df2,left_on='번호',right_on='고객번호')
#df1의 데이터들이 다 나오도록 하는 것이 목적이었기에 8행이 나온다.
```

# [ 연습 예제 ]

```python
staff = pd.DataFrame([{'이름' : '홍길동', '부서' : '영업부'},
                      {'이름' : '임꺽정', '부서' : '개발팀'},
                      {'이름' : '전우치', '부서' : 'R&D'},                    
                      {'이름' : '손오공', '부서' : '개발팀'},
                      {'이름' : '저팔계', '부서' : '영업부'},
                     ])
staff
```

```python
#학력정보 넣어보기
education = pd.DataFrame([{'성명' : '홍길동', '학교' : '가나다대학교', '전공':'컴퓨터'},
                      {'성명' : '임꺽정', '학교' : '나다대학교','전공':'미디어'},
                      {'성명' : '손오공', '학교' : '가다대학교','전공':'통신'},                    
                      {'성명' : '저팔계', '학교' : '가나대학교','전공':'전자정보'},
                      {'성명' : '사오정', '부서' : '파하대학교','전공':'응용미술'},
                     ])
education
```

```python
# 1) 이름 , 학교, 전공, 부서
sample1 = pd.merge(staff,education,left_on='이름',right_on='성명')
sample1
```

```python
# 2)이름, 전공, 부서
sample1.loc[:, ['이름','학교','전공','부서']]
```
